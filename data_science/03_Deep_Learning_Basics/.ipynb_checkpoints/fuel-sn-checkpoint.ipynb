# Basic regression: Predict fuel efficiency 

https://www.tensorflow.org/tutorials/keras/regression

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

# Data
* 자동차 연비 효율 데이터셋 변수
    * Cylinders : 몇기통
    * Displacement : 배기량
    * Horsepower : 마력
    * Weight : 중량
    * Acceleration : 가속
    * Model year : 몇년식
    * Origin (Europe, Japan, USA)
    * MPG : Mile per Gallern, 연비 -> 타겟변수

```python
#  download and import the dataset using pandas:
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data'

column_names = ['MPG', 'Cylinders', 'Displacement', 'Horsepower', 'Weight',
                'Acceleration', 'Model Year', 'Origin']

raw_dataset = pd.read_csv(url, names = column_names, na_values = '?', 
                          comment = '\t', sep = ' ', skipinitialspace = True)
```

```python
dataset = raw_dataset.copy()
dataset.tail()
```

```python
# clean data set: drop missing values
dataset.isna().sum() #각 변수 결측치가 어떤건지 본다.
```

```python
dataset = dataset.dropna() # 결측값 제거해버리기
```

```python
dataset['Origin'] = dataset['Origin'].map({1: 'USA', 2: 'Europe', 3: 'Japan'})

dataset = pd.get_dummies(dataset, prefix = '', prefix_sep = '') # 더미변수 만들기
dataset.tail()
```

```python
# split train and test data
train_dataset = dataset.sample(frac = 0.8, random_state = 0)
test_dataset = dataset.drop(train_dataset.index) # train dataset에 포함된 인덱스를 제외한 dataset을 트레인 데이터셋으로 할당함
```

```python
# Separate the target value, the "label", from the features. 
# This label is the value that you will train the model to predict.
train_features = train_dataset.copy()
test_features = test_dataset.copy()

train_labels = train_features.pop('MPG')
test_labels = test_features.pop('MPG')
```

```python
model = tf.keras.models.Sequential([
  tf.keras.layers.Dense(1, input_shape = (9, )) # 종속변수가 단일변수(1개) 다. 또한, 설명변수인 x가 9개다
])

model.summary()
```

```python
model.compile(loss = 'mean_squared_error', optimizer = tf.keras.optimizers.Adam(0.002), metrics = 'Accuracy') # y와 x가 연속변수인 경우 손실함수는 L2를 많이 쓴다(mean squared error)

# y가 연속변수인 경우 경사하강법으로 최적화시키는 optimizer로 Adam을 쓰는데, 이게 오후에 볼 예정이다. SGD 말고도 딴거 많이 쓸 수 있다. 자세한내용은 오후에~
# learning rate = 0.002
```

```python
# fit the keras model on the dataset
model.fit(train_features, train_labels, epochs = 1000, batch_size = 10)
```

```python
test_results = model.evaluate(test_features, test_labels, verbose = 0) # Loss가 약 13정도이므로, 그렇게 과적합 없다.
test_results
```

```python
test_predictions = model.predict(test_features).flatten()

a = plt.axes(aspect='equal')

plt.scatter(test_labels, test_predictions)
plt.xlabel('True Values [MPG]')
plt.ylabel('Predictions [MPG]')

lims = [0, 50]
plt.xlim(lims)
plt.ylim(lims)

_ = plt.plot(lims, lims)
```
