# Bankrupt data



```python
# Importing the libraries
from numpy import loadtxt
import numpy as np

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.layers.experimental import preprocessing
```

```python
# from google.colab import drive
# drive.mount('/content/drive')
```

```python
data_train_path = 'D:/OneDrive - 연세대학교 (Yonsei University)/Lectures/2022-2_특강/221105_딥러닝기초/dataset/bankruptTrain.csv'
```

```python
# load the dataset
dataset = loadtxt(data_train_path, delimiter = ',')
#dataset = loadtxt('/content/drive/MyDrive/bankruptTrain.csv', delimiter=',')

# split into input (X) and output (y) variables
train_x = dataset[:, 0:4]
train_y = dataset[:, 4]
```

```python
# normalization 
normalizer = preprocessing.Normalization() #표준화
normalizer.adapt(np.array(train_x))

print(normalizer.mean.numpy())
```

```python
model = tf.keras.models.Sequential([ # 어떻게 이 은닉층 레이어를 쌓아서 구조를 만들건지. 정답은 없다. 어떤 네트워크 구조를 만드느냐에 따라 퍼포먼스가 달라진다
    
    normalizer, # 이게 input을 넣을 때, normalizing을 해서 넣겠다는 의미다. input layer에서 모든 인풋들을 각각 표준화한다.
    
    ############## 은닉층 ###############
    tf.keras.layers.Dense(24, activation = 'relu'), # 첫번째 은닉층. 24 - 표준화된 인풋. 4개 변수에 대해 24개 뉴런을 사용한다.(걍 24로 한거다.) 히든레이어는 일단 relu
    layers.Dropout(0.1), # 몇개씩 랜덤하게 탈락시켜서 패스를 없앨거냐 : 과적합 방지용
    
    tf.keras.layers.Dense(24, activation = 'relu'), # 두번째 은닉층. 히든레이어는 일단 relu
    layers.Dropout(0.1), # 과적합 방지용. 0.1 = Dropout rate. 10%의 뉴런을 지워가며 해보겠다는거다
    
    ############## 출력층 ###############
    tf.keras.layers.Dense(1, activation = 'sigmoid') # 결과값이 이진변수니까 로지스틱이다. sigmoid로 활성화함수를 설정해줘야 한다. Output 차원은 1이지.

])

model.summary() #전체 트레이닝 가능한 파라미터가 745개다. 이때 트레인 안되는 파라미터는 9개인데, 이게 normalization 때문에 나온거다. 이건 그냥 인풋 변환한거니까 트레이닝하면 안되지.즉, 745차원에서 최적화 문제를 풀어 나간다.
```

```python
# compile the keras model
model.compile(loss = 'binary_crossentropy', optimizer = keras.optimizers.SGD(0.01), metrics = ['accuracy']) # SGD 사용. 뭐 Adam 써도 되긴하는데.
# fit the keras model on the dataset

model.fit(train_x, train_y, epochs = 1500, batch_size = 10)
```

```python
# evaluate the keras model
_, accuracy = model.evaluate(train_x, train_y, verbose = 1)
print('Accuracy: %.2f' % (accuracy * 100)) # Loss 0.05, Accurady 0.9783으로, 단일 뉴런일 때보다 매우 성능이 높아진 것을 확인할 수 있다.
```
