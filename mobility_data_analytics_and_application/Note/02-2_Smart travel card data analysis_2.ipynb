##### Trip chain analysis

```python
# Mode set
Freq_mode_set = pd.pivot_table(Cdata_C, values = 'Card_ID', index = ['Mode_set'], aggfunc = 'count', margins = True)

Freq_mode_set.T # 이용한 수단 조합별 빈도가 나온다
# 결합된 정보로부터 통행행태를 뽑고, 이를 바탕으로 인사이트를 뽑아내야 한다. 
```

```python
# Mode_set simplification
def Mode_set_text(row): # 이용한 대중교통 리스트를 카테고리 이름으로 정리함. 함수 써서
    
    if row['Mode_set'] == "{'subway'}": #참고: set일 경우는 ""을 제거
        return 'subway'
    
    elif row['Mode_set'] == "{'bus'}":
        return 'bus'
    
    else:
        return 'bus_sub'

Cdata_C['Mode_set_n'] = Cdata_C.apply(lambda row: Mode_set_text(row), axis = 1)
# Mode_set_n 컬럼에 Mode_set_text() 함수를 적용해서 수단을 문자로 만들어 버린다.

Cdata_C
```

```python
# Share by mode set
Freq_mode_set = pd.pivot_table(Cdata_C, values = 'Card_ID', index = ['Mode_set_n'], aggfunc = 'count', margins = True)

mode_share = (Freq_mode_set/Freq_mode_set.loc['All']) * 100

mode_share.T #> 각 Mode Set별 분담률. 지하철만 탄 사람이 54%
```

```python
# Average travel time by mode set
pd.pivot_table(Cdata_C, values = ['In_time','TT_time'], index = ['Mode_set_n'], aggfunc = np.mean).T
```

```python
# Travel time distribution by mode set
sns.displot(data = Cdata_C, x = "TT_time", hue = "Mode_set_n", kind = "kde")
```

```python
# Average transfer time
Cdata_C.Tf_time[Cdata_C["Tf_num"] > 0].mean()
```

```python
# Average transfer time by mode set :: 1번 이상 환승한 사람을 대상으로 환승시간 평균 분석
pd.pivot_table(Cdata_C[Cdata_C["Tf_num"] > 0], values = ['Tf_time'], index = ['Mode_set_n'], aggfunc = np.mean).T
```

### Station data import

```python
Sdata_file = 'Station_code_ADMINC.xlsx' 
Sdata_path = os.path.join(data_dir, Sdata_file)
```

```python
# Import station data
Sdata = pd.read_excel(Sdata_path)

Sdata
# 이 자료는 station_code에 따른 위도, 경도 등을 알려주는 기준자료이다.
# ADMINC :: Administrative District Code(행정구역 코드. 읍면동) >>> 이거 알고 싶으면 딴 .csv 찾기
```

### Boarding and alighting at each station

##### Computing number of boarding and alighting passengers at each station from Cdata

```python
# Boarding passenger
On_ride = pd.pivot_table(Cdata, values = 'Card_ID', index = ['On_station'], aggfunc = 'count')

On_ride.T #> 각 station에서 탑승한 인원 카운팅
```

```python
# Alighting passenger
Off_ride = pd.pivot_table(Cdata, values = 'Card_ID', index = ['Off_station'], aggfunc = 'count')

Off_ride.T
```

##### Merging the number of passengers into Sdata

```python
# Pivot table -> DataFrame & Change column name (For borading)
On_ride = On_ride.reset_index() # On_ride 데이터프레임의 인덱스를 컬럼으로 만들기
On_ride.columns = ["station_code", "On_pass"] # On_ride 데이터프레임의 컬럼 이름 변경

On_ride.T
```

```python
# Pivot table -> DataFrame & Change column name (For alighting)
Off_ride = Off_ride.reset_index()
Off_ride.columns = ["station_code", "Off_pass"]

Off_ride.T
```

```python
# Merging: On_ride & Off_ride :: 승하차 테이블을 join함. 
Ride = pd.merge(On_ride, Off_ride, how = 'outer', on = 'station_code')
Ride.T
```

```python
# Merging: Sdata & Ride :: 승하차 승객수와 정류장 정보를 병합함(join)
S_ride = pd.merge(Sdata, Ride, how = 'outer', on = 'station_code')

S_ride
```

### Passenger load analysis (for a specific line)

```python
# Select line data of 서대문03(Line_ID = 11110641)
My_B_line=B_line[B_line.Line_ID==11110641]
My_B_line
```

```python
### (1) Boarding passenger of the selected line (My_B_Line)
On_ride_MyL = pd.pivot_table(Cdata[Cdata["Route"]==11110641], values='Card_ID', index=['On_station'], aggfunc='count')
On_ride_MyL = On_ride_MyL.reset_index()
On_ride_MyL.columns=["station_code","On_pass_MyL"]

### (2) Allight passenger of the selected line (My_B_Line)
Off_ride_MyL = pd.pivot_table(Cdata[Cdata["Route"]==11110641], values='Card_ID', index=['Off_station'], aggfunc='count')
Off_ride_MyL = Off_ride_MyL.reset_index()
Off_ride_MyL.columns=["station_code","Off_pass_MyL"]

### (3) Merging
Ride_MyL= pd.merge(On_ride_MyL, Off_ride_MyL, how='outer', on='station_code')
S_ride_MyL = pd.merge(Sdata, Ride_MyL, how='outer', on='station_code')

### (4) Key colume matching for merging
My_B_line.rename(columns = {"Station_ID":"station_code"}, inplace=True)

### (5) Merging for line
My_B_line = pd.merge(My_B_line, S_ride_MyL, how='left', on='station_code')

My_B_line
```

```python
# NaN = 0 passenger
My_B_line.On_pass_MyL=My_B_line.On_pass_MyL.fillna(0)
My_B_line
```

```python
# Boarding and aliting at each station
My_B_line[['On_pass_MyL','Off_pass_MyL']].plot.bar(figsize=(15,10))
```

```python
# Boarding and aliting at each station
My_B_line[['On_pass_MyL','Off_pass_MyL']].plot.line(figsize=(15,10))
```

```python
# Boarding and aliting at each station
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(15,10))

ax.plot(My_B_line.On_pass_MyL, label = "boarding")
ax.plot(My_B_line.Off_pass_MyL, label = "alighting")

ax.legend()

#labels = My_B_line["name"].tolist() #Font깨짐 이슈
labels = My_B_line.Station_order

plt.xticks(np.arange(0,len(My_B_line)),labels)
plt.grid(True)

plt.show()
```

```python
# Cumulative boarding and alighting
My_B_line['On_pass_MyL_cum']=0
My_B_line['Off_pass_MyL_cum']=0

My_B_line['On_pass_MyL_cum'].iloc[0]=My_B_line['On_pass_MyL'].iloc[0]
My_B_line['Off_pass_MyL_cum'].iloc[0]=My_B_line['Off_pass_MyL'].iloc[0]


for n in range(1,len(My_B_line)):
    My_B_line['On_pass_MyL_cum'].iloc[n]=My_B_line['On_pass_MyL_cum'].iloc[n-1]+My_B_line['On_pass_MyL'].iloc[n]
    My_B_line['Off_pass_MyL_cum'].iloc[n]=My_B_line['Off_pass_MyL_cum'].iloc[n-1]+My_B_line['Off_pass_MyL'].iloc[n]


My_B_line
```

```python
# Draw cumulative boarding and alighting curve
fig, ax = plt.subplots(figsize=(15,10))

ax.plot(My_B_line.On_pass_MyL_cum, label = "Cumulative boarding")
ax.plot(My_B_line.Off_pass_MyL_cum, label = "Cumulative alighting")

ax.legend()

plt.xticks(np.arange(0,len(My_B_line)),labels)
plt.grid(True)

plt.show()
```

```python
# Passenger volume :: Passenger Loading Graph. 실제 탑승한 탑승객들을 누적 그래프 간 차이로 구한다
My_B_line['Pass_V']=My_B_line.On_pass_MyL_cum-My_B_line.Off_pass_MyL_cum
My_B_line
```
