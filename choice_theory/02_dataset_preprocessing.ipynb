# Dummy Variable 만들기

```python
def timefunc(x):
    """집계시분에 따라 시간범위를 분류하여 주는 함수"""
    
    t = int(x[0:2])
    
    if t >= 0 and t < 3 :
        return '00-03'
    elif t >= 3 and t < 6:
        return '03-06'
    elif t >= 6 and t < 9:
        return '06-09'
    elif t >= 9 and t < 12:
        return '09-12'
    elif t >= 12 and t < 15:
        return '12-15'
    elif t >= 15 and t < 18:
        return '15-18'
    elif t >= 18 and t < 21:
        return '18-21'
    elif t >= 21 and t < 24:
        return '21-24'
    else:
        pass
```
